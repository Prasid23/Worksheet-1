```python
#Problem - 4: Matrix Operations:
'''• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:'''
#1. Prove A.A−1 = I.
import numpy as np

A = np.array([[3, 4], [7, 8]])
A_inv = np.linalg.inv(A)
result = np.dot(A, A_inv)
result = np.round(result, decimals=10)

print("Matrix A:")
print(A)

print("\nInverse of A (A^-1):")
print(A_inv)

print("\nProduct A * A^-1:")
print(result)

#2. Prove AB ̸= BA.
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

AB = np.dot(A, B)
BA = np.dot(B, A)

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nProduct AB:")
print(AB)

print("\nProduct BA:")
print(BA)

#3. Prove (AB)
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

AB = np.dot(A, B)
AB_T = np.transpose(AB)
B_T = np.transpose(B)
A_T = np.transpose(A)
B_T_A_T = np.dot(B_T, A_T)

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nProduct AB:")
print(AB)

print("\nTranspose of AB (AB^T):")
print(AB_T)

print("\nTranspose of B (B^T) and A (A^T) product (B^T * A^T):")
print(B_T_A_T)

#B.
import numpy as np

A = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B = np.array([[-1], [-3], [9]])

A_inv = np.linalg.inv(A)

X = np.dot(A_inv, B)

print("Solution (x, y, z):")
print(X)

#Now: solve the above equation using np.linalg.inv function.
import numpy as np

A = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B = np.array([[-1], [-3], [9]])

A_inv = np.linalg.inv(A)

X = np.dot(A_inv, B)

X_direct = np.linalg.solve(A, B)

det_A = np.linalg.det(A)

print("Solution using inverse method (x, y, z):")
print(X)

print("\nDirect solution using np.linalg.solve (x, y, z):")
print(X_direct)

print("\nDeterminant of A:")
print(det_A)





```
